### 전처리

```python
data = pd.read_csv('data.csv')
```

```python
Y_data = data['diagnosis']
```

```python
Y_data.head()
```

```python
# 'B' -> 0으로 'M' -> 1로 보내주자
Y_data = Y_data.map(lambda x : 0 if x =='B' else x)
Y_data = Y_data.map(lambda x : 1 if x =='M' else x)
```

```python
type(Y_data)
```

```python
Y_data.head(2)
```

```python
# id, diagnosis, 칼럼제거
X_data = data.drop(['id','diagnosis'], axis=1)
```

```python
X_data.head(2)
```

### 텐서플로우로 로지스틱 리그레션을 돌려보자

```python
X = tf.placeholder(tf.float32, shape=[None, 30])
Y = tf.placeholder(tf.float32, shape=[None, 1])
W = tf.Variable(tf.random_normal([30,1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
X.shape
```

```python
Y.shape
```

```python
# hyphthesis는 sigmoid 적용
hypothesis_sigmoid = tf.sigmoid(tf.matmul(X,W)+b)
```

```python
#cost / loss function
cost = -tf.reduce_mean( Y*tf.log(hypothesis_sigmoid) + (1-Y)*tf.log(1-hypothesis_sigmoid))
```

```python
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
# Accuracy Computation
# True if hypothersis > 0.5 else False

predicted = tf.cast(hypothesis_sigmoid > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
```

```python
with tf.Session() as sess :
    
    sess.run(tf.global_variables_initializer())
    feed = {X:X_data, Y:Y_data}
    
    for step in range(1001):
        sess.run(train, feed_dict=feed)
        #if step % 200 == 0:
            #print(step, sess.run(cost, feed_dict=feed))
            
    h, c, a = sess.run([hypothesis_sigmoid, predicted, accuracy], feed_dict=feed)
    print("\nHypothesis:",h,"\nCorrect (Y):",c,"\nAccuracy:",a)
```
